```python
def addspliteddate(dataset):
    data_split = {}
    data_split['startyear'] = []
    data_split['startmonth'] = []
    data_split['startday'] = []
    data_split['starthour'] = []
    #data_split['startmin'] = []
    data_split['endyear'] = []
    data_split['endmonth'] = []
    data_split['endday'] = []
    data_split['endhour'] = []
    #data_split['endmin'] = []
    for i in range(len(dataset)):
        #print(pd.to_datetime(dataset.iloc[i].start_date).year)
        data_split['startyear'].append(pd.to_datetime(dataset.iloc[i].start_date).year)
        data_split['startmonth'].append(pd.to_datetime(dataset.iloc[i].start_date).month)
        data_split['startday'].append(pd.to_datetime(dataset.iloc[i].start_date).day)
        data_split['starthour'].append(pd.to_datetime(dataset.iloc[i].start_date).hour)
        #data_split['startmin'].append(pd.to_datetime(dataset.iloc[i].start_date).minute)
        data_split['endyear'].append(pd.to_datetime(dataset.iloc[i].end_date).year)
        data_split['endmonth'].append(pd.to_datetime(dataset.iloc[i].end_date).month)
        data_split['endday'].append(pd.to_datetime(dataset.iloc[i].end_date).day)
        data_split['endhour'].append(pd.to_datetime(dataset.iloc[i].end_date).hour)
        #data_split['endmin'].append(pd.to_datetime(dataset.iloc[i].end_date).minute)
    date_pd = pd.DataFrame(data_split)
    date_pd = date_pd[['startyear', 'startmonth', 'startday', 'starthour',
                      'endyear', 'endmonth', 'endday', 'endhour']]
    return date_pd
```

```python
def splitdate(dataset):
    data_split = {}
    data_split['year'] = []
    data_split['month'] = []
    data_split['day'] = []
    for i in range(len(dataset)):
        #print(i)
        #print(pd.to_datetime(dataset.iloc[i].start_date).year)
        data_split['year'].append(pd.to_datetime(dataset.iloc[i].date).year)
        data_split['month'].append(pd.to_datetime(dataset.iloc[i].date).month)
        data_split['day'].append(pd.to_datetime(dataset.iloc[i].date).day)
    date_pd = pd.DataFrame(data_split)
    date_pd = date_pd[['year', 'month', 'day']]
    return date_pd
```

```python
def match_weather(dataset, weather_dataset):
    data_weather = {}
    data_weather['Temperature'] = []
    data_weather['Dew Point'] = []
    data_weather['Humidity'] = []
    data_weather['Pressure'] = []
    data_weather['Visibility'] = []
    for i in range(len(dataset)):
        #print(i)
        temp = dataset.iloc[i]
        #print(temp)
        #temperature_list = []
        #hour_difference = ((pd.to_datetime(dataset.iloc[i].end_date)-pd.to_datetime(dataset.iloc[i].start_date))/np.timedelta64(1, 'h')).astype(int)
        #for j in range(hour_difference):
        if ((temp.startyear == temp.endyear) & (temp.startmonth == temp.endmonth) & 
            (temp.startday == temp.endday) & (temp.starthour == temp.endhour)):
            #print(1)
            temp_weather = weather_dataset[(weather_dataset.year == temp.startyear) & (weather_dataset.month == temp.startmonth)
                                          & (weather_dataset.day == temp.startday) & (weather_dataset.hour == temp.starthour)
                                          & (weather_dataset.code == temp.city)]
            #print(temp_weather)
            data_weather['Temperature'].append(temp_weather['Temp'].mean())
            data_weather['Dew Point'].append(temp_weather['Dew Point'].mean())
            data_weather['Humidity'].append(temp_weather['Humidity'].mean())
            data_weather['Pressure'].append(temp_weather['Pressure'].mean())
            data_weather['Visibility'].append(temp_weather['Visibility'].mean())
            #condition = temp_weather['Condition'].mean()
        else:
            temp_weather = weather_dataset[(weather_dataset.year == temp.startyear) & (weather_dataset.month == temp.startmonth)
                                          & (weather_dataset.day == temp.startday) & (weather_dataset.hour == temp.starthour)
                                          & (weather_dataset.code == temp.city)]
            start_temperature = temp_weather['Temp'].mean()
            start_dewpoint = temp_weather['Dew Point'].mean()
            start_humidity = temp_weather['Humidity'].mean()
            start_pressure = temp_weather['Pressure'].mean()
            start_visibility = temp_weather['Visibility'].mean()

            temp_weather = weather_dataset[(weather_dataset.year == temp.endyear) & (weather_dataset.month == temp.endmonth)
                                          & (weather_dataset.day == temp.endday) & (weather_dataset.hour == temp.endhour)
                                          & (weather_dataset.code == temp.city)]
            end_temperature = temp_weather['Temp'].mean()
            end_dewpoint = temp_weather['Dew Point'].mean()
            end_humidity = temp_weather['Humidity'].mean()
            end_pressure = temp_weather['Pressure'].mean()
            end_visibility = temp_weather['Visibility'].mean()
            
            if np.isnan(start_temperature) & np.isnan(end_temperature):
                data_weather['Temperature'].append(start_temperature)
                data_weather['Dew Point'].append(start_dewpoint)
                data_weather['Humidity'].append(start_humidity)
                data_weather['Pressure'].append(start_pressure)
                data_weather['Visibility'].append(start_visibility)
            elif np.isnan(start_temperature):
                data_weather['Temperature'].append(end_temperature)
                data_weather['Dew Point'].append(end_dewpoint)
                data_weather['Humidity'].append(end_humidity)
                data_weather['Pressure'].append(end_pressure)
                data_weather['Visibility'].append(end_visibility)
            elif np.isnan(end_temperature):
                data_weather['Temperature'].append(start_temperature)
                data_weather['Dew Point'].append(start_dewpoint)
                data_weather['Humidity'].append(start_humidity)
                data_weather['Pressure'].append(start_pressure)
                data_weather['Visibility'].append(start_visibility)
            else:
                data_weather['Temperature'].append((start_temperature + end_temperature)/2)
                data_weather['Dew Point'].append((start_dewpoint + end_dewpoint)/2)
                data_weather['Humidity'].append((start_humidity + end_humidity)/2)
                data_weather['Pressure'].append((start_pressure + end_pressure)/2)
                data_weather['Visibility'].append((start_visibility + end_visibility)/2)
                
    date_pd = pd.DataFrame(data_weather)
    return date_pd
            
```

```python
def findmissing(dataset, weather_dataset):
    data_weather = {}
    data_weather['Temperature'] = []
    data_weather['Dew Point'] = []
    data_weather['Humidity'] = []
    data_weather['Pressure'] = []
    data_weather['Visibility'] = []
    for i in range(len(dataset)):
        print(i)
        temp = dataset.iloc[i]
        #print(temp)
        
        newstarthour = temp.starthour - 1
        newstartday = temp.startday
        while True:
            #print(newstartday, newstarthour)
            if newstarthour<0:
                newstarthour = newstarthour%24
                newstartday = newstartday-1
            
            temp_weather = weather_dataset[(weather_dataset.year == temp.startyear) & (weather_dataset.month == temp.startmonth)
                                          & (weather_dataset.day == newstartday) & (weather_dataset.hour == newstarthour)
                                          & (weather_dataset.code == temp.city)]
            if not np.isnan(temp_weather['Temp'].mean()):
                break
            elif newstartday < 1:
                break
            else:
                newstarthour -= 1
        #print(1)
        
        start_temperature = temp_weather['Temp'].mean()
        start_dewpoint = temp_weather['Dew Point'].mean()
        start_humidity = temp_weather['Humidity'].mean()
        start_pressure = temp_weather['Pressure'].mean()
        start_visibility = temp_weather['Visibility'].mean()
        
        newendhour = temp.endhour + 1
        newendday = temp.endday
        while True:
            if newendhour > 23:
                newendhour = newendhour%24
                newendday += 1
                
            temp_weather = weather_dataset[(weather_dataset.year == temp.endyear) & (weather_dataset.month == temp.endmonth)
                                          & (weather_dataset.day == newendday) & (weather_dataset.hour == newendhour)
                                          & (weather_dataset.code == temp.city)]
            if not np.isnan(temp_weather['Temp'].mean()):
                break
            elif newendday > 31:
                break
            else:
                newendhour += 1
        #print(2)
        
        end_temperature = temp_weather['Temp'].mean()
        end_dewpoint = temp_weather['Dew Point'].mean()
        end_humidity = temp_weather['Humidity'].mean()
        end_pressure = temp_weather['Pressure'].mean()
        end_visibility = temp_weather['Visibility'].mean()
        
        
        if np.isnan(start_temperature) & np.isnan(end_temperature):
            data_weather['Temperature'].append(start_temperature)
            data_weather['Dew Point'].append(start_dewpoint)
            data_weather['Humidity'].append(start_humidity)
            data_weather['Pressure'].append(start_pressure)
            data_weather['Visibility'].append(start_visibility)
        elif np.isnan(start_temperature):
            data_weather['Temperature'].append(end_temperature)
            data_weather['Dew Point'].append(end_dewpoint)
            data_weather['Humidity'].append(end_humidity)
            data_weather['Pressure'].append(end_pressure)
            data_weather['Visibility'].append(end_visibility)
        elif np.isnan(end_temperature):
            data_weather['Temperature'].append(start_temperature)
            data_weather['Dew Point'].append(start_dewpoint)
            data_weather['Humidity'].append(start_humidity)
            data_weather['Pressure'].append(start_pressure)
            data_weather['Visibility'].append(start_visibility)
        else:
            data_weather['Temperature'].append((start_temperature + end_temperature)/2)
            data_weather['Dew Point'].append((start_dewpoint + end_dewpoint)/2)
            data_weather['Humidity'].append((start_humidity + end_humidity)/2)
            data_weather['Pressure'].append((start_pressure + end_pressure)/2)
            data_weather['Visibility'].append((start_visibility + end_visibility)/2)
        
    date_pd = pd.DataFrame(data_weather)
    return date_pd
            
```

```python
def updateweather(ori_dataset, weather):
    for i in range(len(weather)):
        print(i)
        temp = weather.iloc[i]
        update_id = temp.id
        ori_dataset.loc[ori_dataset['id']==update_id, 'Temperature'] = temp.Temperature
        ori_dataset.loc[ori_dataset['id']==update_id, 'Dew Point'] = temp['Dew Point']
        ori_dataset.loc[ori_dataset['id']==update_id, 'Humidity'] = temp.Humidity
        ori_dataset.loc[ori_dataset['id']==update_id, 'Pressure'] = temp.Pressure
        ori_dataset.loc[ori_dataset['id']==update_id, 'Visibility'] = temp.Visibility
    return ori_dataset
```
